```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df = pd.read_csv("./pokemon-images-and-types/pokemon.csv")
df1 = pd.read_csv("./pokemon/pokemon_alopez247.csv")
```

```python
names = df.iloc[:,0]
names = [name.lower() for name in names]
```

```python
names1 = df1.iloc[:,1].values
names1 = [name.lower() for name in names1]
```

```python
df.iloc[:, 0] = names
df1.iloc[:, 1] = names1
```

```python
dff = pd.merge(df, df1, how='inner', on='Name')
```

```python
mdf = dff.drop(['Type_1','Type_2'], axis=1)
```
